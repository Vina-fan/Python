### 对性别分析

```python
# 做简单的汇总统计（常用）
titanic.Sex.value_counts()
```

```python
# 生还者中，男女的人数
titanic[titanic.Survived == 1].Sex.value_counts()
```

```python
# 未生还者中，男女的人数
titanic[titanic.Survived==0].Sex.value_counts()
```

```python
survived = titanic[titanic.Survived==1].Sex.value_counts()
dead = titanic[titanic.Survived==0].Sex.value_counts()
df = pd.DataFrame([survived,dead],index=['survived','dead'])
df
```

```python
df.plot.bar()
```

```python
# 把dataframe转置
df = df.T  
df
```

```python
df.plot.bar()#等于df.plot(kind='bar')
```

```python
df.plot(kind='bar',stacked=True)
```

```python
df.sum()
```

```python
#不同性别生还者的比例情况
df['p_survived']=df.survived/(df.survived+df.dead)
df['p_dead']=df.dead/(df.survived+df.dead)
df[['p_survived','p_dead']].plot.bar(stacked=True)
```

由上图可以看出：性别对生还的影响较大

###  对年龄分析

- 简单统计

```python
titanic.Age.value_counts()
```

```python
survived=titanic[titanic.Survived==1].Age
dead=titanic[titanic.Survived==0].Age
df = pd.DataFrame([survived,dead],index=['survived','dead'])
df = df.T
df
```

```python
df.plot.hist(stacked=True)
```

```python
df.plot.hist(stacked=True,bins=30)
# 中间频率较高是由于空值全部替换为了中位数
```

```python
# 密度图
df.plot.kde()
```

```python
titanic.Age.describe()#查看极值，确定取值范围
```

```python
# 限定范围
df.plot.kde(xlim=(0,80))
```

```python
age = 16
young = titanic[titanic.Age<=age]['Survived'].value_counts()
old = titanic[titanic.Age>age]['Survived'].value_counts()
df = pd.DataFrame([young,old],index=['young','old'])
df.columns = ['dead','survived']
df.plot.bar(stacked=True)
```

```python
#不同性别生还者的比例情况
df['p_survived']=df.survived/(df.survived+df.dead)
df['p_dead']=df.dead/(df.survived+df.dead)
df[['p_survived','p_dead']].plot.bar(stacked=True)
```

### 组合特征

```python
import matplotlib.pyplot as plt
plt.scatter(titanic[titanic.Survived==0].Age,titanic[titanic.Survived==0].Fare)
```

```python
# 未生还者
ax = plt.subplot()
age = titanic[titanic.Survived==0].Age
fare = titanic[titanic.Survived==0].Fare
plt.scatter(age,fare,s=10,marker='*',alpha=0.3,linewidths=3,edgecolors='gray')
ax.set_xlabel('age')
ax.set_ylabel('fare')
```

```python
# 生还者
ax = plt.subplot()
age = titanic[titanic.Survived==1].Age
fare = titanic[titanic.Survived==1].Fare
plt.scatter(age,fare,s=10,marker='*',alpha=0.3,linewidths=3,edgecolors='red')
ax.set_xlabel('age')
ax.set_ylabel('fare')
```

```python
# 在同一张图中进行对比

# 未生还者（灰色）
ax = plt.subplot()
age = titanic[titanic.Survived==0].Age
fare = titanic[titanic.Survived==0].Fare
plt.scatter(age,fare,s=10,marker='*',alpha=0.3,linewidths=3,edgecolors='gray')
ax.set_xlabel('age')
ax.set_ylabel('fare')

# 生还者（红色）
ax = plt.subplot()
age = titanic[titanic.Survived==1].Age
fare = titanic[titanic.Survived==1].Fare
plt.scatter(age,fare,s=10,marker='*',alpha=0.3,linewidths=3,edgecolors='red')
ax.set_xlabel('age')
ax.set_ylabel('fare')
```
